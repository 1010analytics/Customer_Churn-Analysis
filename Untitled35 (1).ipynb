```python
from sklearn.model_selection import train_test_split
```

```python
x= copy_data.drop('Churn ', axis=1)
```

```python
y= copy_data['Churn ']
```

```python
x_tr, x_ts,y_tr,y_ts = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

models = [LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), KNeighborsClassifier()]
```

```python
for model in models:
  model.fit(x_tr, y_tr)
  y_pred = model.predict(x_ts)
  #accuracy = accuracy_score(y_ts, y_pred)
  ma = mean_absolute_error(y_ts, y_pred)
  mse= mean_squared_error(y_ts, y_pred)
  rmse= np.sqrt(mse)
  r2= r2_score(y_ts, y_pred)
  print(model,":-")
  print( "Mean Absolute Error of", model, "is=", ma)
  print("Means Squared Error of", model, "is=", mse)
  print("Root mean Squared Error of", model, "is=", rmse)
  print("r2 score of", model, "is=",r2)
  print()
```
